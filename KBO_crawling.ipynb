```python
import time
```

```python
column = ["DATE","상대","AVG1","PA","AB","R","H","2B","3B","HR","RBI","SB","CS","BB","HBP","SO","GDP"]
df = pd.DataFrame(columns=column)

for pcode in pcode_list:
  url = "https://www.koreabaseball.com/Record/Player/HitterDetail/Daily.aspx?playerId={playerId}".format(playerId = pcode)
  print(url)
  chromeDriver.get(url) 
  html = chromeDriver.page_source 
  bsObject = BeautifulSoup(html, 'html.parser')

  templen = len(bsObject.find_all('tr'))
  print(templen)
  for i in range(2, templen):
    tempTr = bsObject.find_all("tr")[i]
    if (tempTr.find("th") is not None):
      continue
    row = {}
    column_idx = 0
    for j in range(17):
      tempTd = tempTr.find_all("td")[j].text
      if (tempTd == "" and j > 0):
        continue
      if j == 0:
        tempTd = tempTd + ".21"
      row[column[column_idx]] = tempTd
      column_idx += 1
    row['PCODE'] = int(pcode)
    df = df.append(row, ignore_index=True)

  click_on_button = chromeDriver.find_element_by_xpath('//*[@id="cphContents_cphContents_cphContents_ddlYear"]')
  click_on_button.click()
  click_on_button = chromeDriver.find_element_by_xpath('//*[@id="cphContents_cphContents_cphContents_ddlYear"]/option[2]')
  click_on_button.click()
  time.sleep(2)
  html = chromeDriver.page_source 
  bsObject = BeautifulSoup(html, 'html.parser')
  templen = len(bsObject.find_all('tr'))
  print(templen)
  for i in range(2, templen):
    tempTr = bsObject.find_all("tr")[i]
    if (tempTr.find("th") is not None):
      continue
    row = {}
    column_idx = 0
    for j in range(17):
      tempTd = tempTr.find_all("td")[j].text
      if (tempTd == "" and j > 0):
        continue
      if j == 0:
        tempTd = tempTd + ".20"
      row[column[column_idx]] = tempTd
      column_idx += 1
    row['PCODE'] = int(pcode)
    df = df.append(row, ignore_index=True)

  click_on_button = chromeDriver.find_element_by_xpath('//*[@id="cphContents_cphContents_cphContents_ddlYear"]')
  click_on_button.click()
  click_on_button = chromeDriver.find_element_by_xpath('//*[@id="cphContents_cphContents_cphContents_ddlYear"]/option[3]')
  click_on_button.click()
  time.sleep(2)
  html = chromeDriver.page_source 
  bsObject = BeautifulSoup(html, 'html.parser')
  templen = len(bsObject.find_all('tr'))
  print(templen)
  for i in range(2, templen):
    tempTr = bsObject.find_all("tr")[i]
    if (tempTr.find("th") is not None):
      continue
    row = {}
    column_idx = 0
    for j in range(17):
      tempTd = tempTr.find_all("td")[j].text
      if (tempTd == "" and j > 0):
        continue
      if j == 0:
        tempTd = tempTd + ".19"
      row[column[column_idx]] = tempTd
      column_idx += 1
    row['PCODE'] = int(pcode)
    df = df.append(row, ignore_index=True)

  click_on_button = chromeDriver.find_element_by_xpath('//*[@id="cphContents_cphContents_cphContents_ddlYear"]')
  click_on_button.click()
  click_on_button = chromeDriver.find_element_by_xpath('//*[@id="cphContents_cphContents_cphContents_ddlYear"]/option[4]')
  click_on_button.click()
  time.sleep(2)
  html = chromeDriver.page_source 
  bsObject = BeautifulSoup(html, 'html.parser')
  templen = len(bsObject.find_all('tr'))
  print(templen)
  for i in range(2, templen):
    tempTr = bsObject.find_all("tr")[i]
    if (tempTr.find("th") is not None):
      continue
    row = {}
    column_idx = 0
    for j in range(17):
      tempTd = tempTr.find_all("td")[j].text
      if (tempTd == "" and j > 0):
        continue
      if j == 0:
        tempTd = tempTd + ".18"
      row[column[column_idx]] = tempTd
      column_idx += 1
    row['PCODE'] = int(pcode)
    df = df.append(row, ignore_index=True)
df


```
